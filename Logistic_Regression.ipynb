***X & y***

```python
X=df.drop('Outcome', axis=1)
y=df['Outcome']
```

***Train Test Split***

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2, random_state=16) 
```

# Modelling
- logistic regression baseline model

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
log_reg.fit(X_train,y_train)
```

```python
log_reg.intercept_
```

```python
log_reg.coef_
```

# Evaluation

```python
# predict train data
ypred_train=log_reg.predict(X_train)

from sklearn.metrics import accuracy_score
print('Train Accuracy score is : ', accuracy_score(y_train, ypred_train))
```

```python
# CV score

from sklearn.model_selection import cross_val_score
print('cross validation score is : ', cross_val_score(log_reg, X_train, y_train, cv=5, scoring='accuracy').mean())
```

```python
# predict test data
ypred_test=log_reg.predict(X_test)

from sklearn.metrics import accuracy_score
print('Test Accuracy score is : ', accuracy_score(y_test, ypred_test))
```

***Confusion Matrix***

```python
from sklearn.metrics import confusion_matrix

cm= confusion_matrix(y_test, ypred_test)
cm
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay(cm, display_labels=[0,1]).plot()
plt.show()
```

***Classification Report***

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, ypred_test))
```

***ROC Curve***

```python
from sklearn.metrics import roc_curve, auc, RocCurveDisplay

fpr, tpr, thresholds = roc_curve(y_test, ypred_test)

roc_auc=auc(fpr, tpr)

RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc).plot()

plt.show()
```

***AUC Score***

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test, ypred_test)
```
